```python
#loading the dataset
import pandas as pd

df = pd.read_csv("Cross_border.csv")

print(df.head())
print(df.shape)
print(df.info())
```

```python
#descriptive statistics   this will summarize the dataset
print(df.describe())
```

```python
#statistics summarize on the  transaction
print(df["Amount"].describe())
```

```python
#checking for the missing value
print(df.isnull().sum())
```

```python
#fraud class distribution
print(df["Is_laundering"].value_counts())

#  0->legitimate
#  1->fraud
```

```python
#percentage distribution
print(df["Is_laundering"].value_counts(normalize=True)*100)

```

```python
#visualize fraud distribution
import matplotlib.pyplot as plt

df["Is_laundering"].value_counts().plot(kind="bar")

plt.title("Fraud vs Legitimate Transactions")
plt.xlabel("Transaction Type")
plt.ylabel("Count")
plt.show()
```

```python
#analysing transaction amount 
import seaborn as sns

sns.boxplot(x="Is_laundering", y="Amount", data=df)

plt.title("Transaction Amount vs Fraud")
plt.show()
```
