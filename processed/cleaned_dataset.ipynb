```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load data
df = pd.read_csv(r"C:\Users\rchan\Downloads\SAML-D.csv\reduced_dataset.csv")

df = df.dropna()

df = df.drop_duplicates()

# Handle Outliers for 'Amount' using 1st and 99th percentiles (clipping)
q_low = df["Amount"].quantile(0.01)
q_hi  = df["Amount"].quantile(0.99)
df["Amount"] = df["Amount"].clip(lower=q_low, upper=q_hi)

# Encode categorical data
# Let's use LabelEncoder for simplicity
cat_cols = ['Payment_currency', 'Received_currency', 'Sender_bank_location', 
            'Receiver_bank_location', 'Payment_type', 'Laundering_type']
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))

# Convert Date and Time to standard formats, or let's leave them or drop them depending on general use.
# It's better to keep them but maybe just encode or keep as is. Let's just keep as is for Date/Time.

# Save the cleaned dataset
output_file = 'cleaned_dataset.csv'
df.to_csv(output_file, index=False)
print("File saved as", output_file)
```
